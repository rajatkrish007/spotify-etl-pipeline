<h3>Data extraction and transformation (From Json to dataframe)</h3>

```python
# Extracting values from the list named items

data['items'][0]['track']
```

```python
# Extracting first track detail

data['items'][0]['track']
```

```python
# Extracting first album

data['items'][0]['track']['album']
```

```python
# Extracting track details from album

data['items'][1]['track']['album']['id']
data['items'][1]['track']['album']['name']
data['items'][1]['track']['album']['type']
data['items'][1]['track']['album']['total_tracks']
data['items'][1]['track']['album']['release_date']
data['items'][1]['track']['album']['external_urls']['spotify']
```

```python
# Now similarly we will be extracting details from all the albums and creating the extracted values as dictionary to convert it into dataframe

album_list = []
for album in data['items']:
    album_dict = {
    'album_id': album['track']['album']['id'],
    'album_name': album['track']['album']['name'],
    'album_total_tracks': album['track']['album']['total_tracks'],
    'album_release_date': album['track']['album']['release_date'],
    'album_url': album['track']['album']['external_urls']['spotify']
    }

    album_list.append(album_dict)
```

```python
album_list
```

```python
import pandas as pd
```

```python
album_df = pd.DataFrame(album_list)
album_df
```

```python
data['items']
```

```python
# Similarly we will be extracting data for Artists

data['items'][0]['track']['album']['artists']
# data['items'][1]['track']['artists']
# data['items'][2]['track']['artists']
```

```python
artist_list = []
for art in data['items']:
    for key, value in art.items():
        if key == 'track':
            for art in value['artists']:
                artist_elems = {
                    'artist_id': art['id'],
                    'artist_name': art['name'],
                    'external_link': art['href']
                
                }
    artist_list.append(artist_elems)
```

```python
# Converting this data into dataframe

artist_list
```

```python
# Converting it to dataframe

artist_df = pd.DataFrame(artist_list)
artist_df
```

```python
data['items']
```

```python
# Same for songs 

song_list = []

for songs in data['items']:

    song_id = songs['track']['id']
    song_name = songs['track']['name']
    song_duration = songs['track']['duration_ms']
    song_url = songs['track']['external_urls']['spotify']
    song_popularity = songs['track']['popularity']
    album_id = songs['track']['album']['id']
    artist_id = songs['track']['album']['artists'][0]['id']
    
    song_dict = {
    'song_id': song_id,
    'song_name': song_name,
    'song_duration': song_duration,
    'song_url': song_url,
    'song_popularity': song_popularity,
    'album_id': album_id,
    'artist_id': artist_id,
    }

    song_list.append(song_dict)
```

```python
song_list
```

```python
# Converting the list into dataframe

song_df = pd.DataFrame(song_list)
```
